```python
#Looking at population vs vaccinations in canada at a daily level

pysqldf("select cd.location, cd.date , cd.total_cases, cd.new_cases, cd.total_deaths, cd.new_deaths, cv.new_vaccinations, cv.total_vaccinations from covid_deaths cd join covid_vaccinations cv on cd.continent = cv.continent and cd.location = cv.location and cd.date = cv.date where lower(cd.location) = 'canada' limit 10;")
```

Now i will modify the above query so that i can do a rolling count instead of using a total vaccinations column, i will be performing operations on new vaccinations column in order to get a total vaccinations column.

```python
pysqldf("select cd.location, cd.date , cd.total_cases, cd.new_cases, cd.total_deaths, cd.new_deaths, cv.new_vaccinations, sum(cast(cv.new_vaccinations as int)) OVER (partition by cd.location order by cd.location, cd.date) as rolling_sum_vaccinations from covid_deaths as cd join covid_vaccinations as cv on cd.continent = cv.continent and cd.location = cv.location and cd.date = cv.date where lower(cd.location) = 'canada' order by cd.date, rolling_sum_vaccinations limit 10;")
```

For a CTE to work, the number of output columns in a CTE should match the input columns inside the CTE query 

```python
pysqldf("with populationvsvaccine (continent, location, date, population, new_vaccinations, rolling_sum_vaccinated) as (select cd.continent, cd.location, cd.date , cd.population, cv.new_vaccinations, sum(cast(cv.new_vaccinations as int)) OVER (partition by cd.location order by cd.location, cd.date) as rolling_sum_vaccinations from covid_deaths as cd join covid_vaccinations as cv on cd.continent = cv.continent and cd.location = cv.location and cd.date = cv.date where lower(cd.location) = 'canada')select * from populationvsvaccine order by date limit 10;")
```

```python
pysqldf("with populationvsvaccine (continent, location, date, population, new_vaccinations, rolling_sum_vaccinated) as (select cd.continent, cd.location, cd.date , cd.population, cv.new_vaccinations, sum(cast(cv.new_vaccinations as int)) OVER (partition by cd.location order by cd.location, cd.date) as rolling_sum_vaccinations from covid_deaths as cd join covid_vaccinations as cv on cd.continent = cv.continent and cd.location = cv.location and cd.date = cv.date where lower(cd.location) = 'canada')select *,(rolling_sum_vaccinated/population)*100 as roll from populationvsvaccine order by date limit 10;")
```

##In order to make the queries run a bit faster, i am going to create a subset of the dataset and perform operations on it, so for that i will be creating a  **temp table** in the next line and reference it for future operations. *italicized text*

```python
pysqldf("drop table if exists #popvac create table #popvac (continent nvarchar(255), location nvarchar(255), date datetime, population numeric, new_vaccinations numeric, rolling_sum_vaccinated numeric) insert into #popvac select cd.continent, cd.location, cd.date , cd.population, cv.new_vaccinations, sum(cast(cv.new_vaccinations as int)) OVER (partition by cd.location order by cd.location, cd.date) as rolling_sum_vaccinations from covid_deaths as cd join covid_vaccinations as cv on cd.continent = cv.continent and cd.location = cv.location and cd.date = cv.date where lower(cd.location) = 'canada' select * from popvac;")
```
