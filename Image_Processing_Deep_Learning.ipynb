```python
#Resizing the image using Pillow Library
img2 = Image.open('/content/red-fire-hydrant-with-white-background-picture-id188026946?k=20&m=188026946&s=612x612&w=0&h=RbtVoYuuA5aw93dBavHvNjAAjtWwzoa4akmokjSNsXQ=')
img_resize = img2.resize((150,150))
img_resize.save('hydrant_resize.jpg')
```

```python
#converting rgb images to grayscale using openCV
img3 = cv2.imread('/content/hydrant_resize.jpg')
```

```python
#Checking the data type of the resized image
type(img3)
```

```python
#Checking the dimensions of the resized image
img3.shape
```

```python
#Converting the resized image into a grayscale image for reduced memory usage
gray_img3 = cv2.cvtColor(img3, cv2.COLOR_RGB2GRAY)
```

```python
#Saving the image after resizing and converting it to grayscale as a new image
cv2.imwrite('hydrant_gray.jpg', gray_img3)
```
