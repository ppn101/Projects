```python
import numpy as np #To Create Numpy Arrays
import pandas as pd #To create Dataframe
import sklearn #Since i will be creating a Logistic regression model for binary classification hence, i will be using Sklearn for the same
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression #Here i have specified that i will be using logistic regression to solve this problem
from sklearn.metrics import accuracy_score #Since at the last i need to identify the accuracy of my model hence, i have imported the required function to check the same.
```

```python
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Sonar data.csv", header= None) #Importing the file as a dataframe
```

```python
df[60].value_counts() #This will help with grouping the values with their counts in the classifier row. Here M refers to Mine and R refers to Rock
```

```python
df.groupby(60).mean() #Since there is visible difference between the values observed for statistical metrics for metal and rock hence taking out mean served to be helpful here.
```

```python
#Since this is a supervised learning model hence i will be using both features and labels to train the model.
feature = df.drop(columns = 60, axis = 1) # This will create a dataframe without the last classified data column, axis has been set to 1 so as to instruct python to drop the column and not the row.
label = df[60]
```

```python
label.head(3)
```

```python
#First i will split the data of both feature and label into training data and test data.
#Then i will be selecting a test sample size to distribute it between feature_test and label_test.
#Then i will instruct the model to do the data segregation in terms of the label values i.e M & R.
#Then i will initialise the random_state to 1 in order to attain result repoducibility.

feature_train, feature_test, label_train, label_test = train_test_split(feature, label, test_size = 0.05, stratify = label, random_state = 1)
```

```python
print(feature.shape, feature_test.shape, feature_train.shape) #I will run this code to get a sense of how the splitting has been done by python according to the instructions provided.
```

```python
model = LogisticRegression() #This will load the logistic regression model into the variable model
```

```python
model.fit(feature_train, label_train)
```

```python
#On a general note - Accuracy will always be higher for training data as compared to test data since the model has already seen the training data.
#As the no. of data points increase, the model take more computational resources to construct.

feature_train_predict = model.predict(feature_train)
training_accuracy = accuracy_score(feature_train_predict, label_train)
```

```python
#Now i will computre the accuracy of the model for the test data
feature_test_predict = model.predict(feature_test)
testing_accuracy = accuracy_score(feature_test_predict, label_test)
```

```python
#Now i will print out the accuracy scores for train and test data in order to check model health and to establish if the model needs more tuning.
print(round(training_accuracy,2)*100)
print(round(testing_accuracy,2)*100)
```
