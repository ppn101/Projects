```python
#Importing the required libraries and functions
import numpy as np
import pandas as pd
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
#Importing the dataset and storing it as a dataframe for further analysis
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/breast_cancer_data.csv")
```

```python
#Printing the first 3 rows in the dataframe
df.head(3)
```

```python
#checking the number of columns and rows in the dataframe
df.shape
```

```python
#Checking the count of null rows in the entire dataframe - columnwise
df.isnull().sum()
```

```python
#Getting info on the column datatypes and count of null values.
df.info
```

```python
#Getting a statistical summary of the entire dataframe
df.describe()
```

```python
#Getting the categorical count of values in the unnamed column
df["Unnamed: 32"].value_counts()
```

```python
#Deleting the column from our dataframe since it is not a significant variable for our predictions set
df.drop(columns= 'Unnamed: 32', axis = 1, inplace = True)
```

```python
#Getting the categorical count of values in the diagnosis column in order to see the split of values
df['diagnosis'].value_counts()
```

```python
#Since the diagnosis column is our target variable hence we will output the mean as a statistical differentiator for verifying if there is a scope of model training to predict the target variable
df.groupby('diagnosis').mean()
```

```python
#Since textual values in a column cannot be fed into the model, hence i will be replacing them with numerical values.
df['diagnosis'].replace(to_replace = 'B', value = 0, inplace = True)
df['diagnosis'].replace(to_replace = 'M', value = 1, inplace = True)
```

```python
#Creating a dataframe for target variable
label = df['diagnosis']
```

```python
label.shape
```

```python
#Creating a dataframe for feature set
features = df.drop(columns = ['id','diagnosis'], axis = 1)
```

```python
features.shape
```

```python
features.head(3)
```

```python
#Splitting the features and label dataframes into train and test dataframes for training and testing the model
features_train, features_test, label_train, label_test = train_test_split(features, label, test_size= 0.2, stratify = label, random_state = 1)
```

```python
print(features_train.shape)
print(features_test.shape)
print(label_train.shape)
print(label_test.shape)
```

```python
#Selecting logistic regression as the model type for (use case: classification) - performing the analysis
classifier = LogisticRegression()
```

```python
#Training the model on the input training data of features and corresponding labels
classifier.fit(features_train, label_train)
```

```python
#calculating the accuracy of the mdodel on the training data
predict_train = classifier.predict(features_train)
accuracy_train = accuracy_score(predict_train, label_train)
print(round(accuracy_train,2)*100)
```

```python
#calculating the accuracy of the mdodel on the test data
predict_test = classifier.predict(features_test)
accuracy_test = accuracy_score(predict_test, label_test)
print(round(accuracy_test,2)*100)
```
