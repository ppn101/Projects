```python
#Importing all the required libraries and functions.
import numpy as np
import pandas as pd
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
#Importing the file and storing it as a dataframe.
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/creditcard.csv")
```

```python
#Returning the first 3 rows of the dataframe.
df.head(3)
```

```python
#Checking the number of rows and columns in the dataframe.
df.shape
```

```python
#Obtaining a statistical summary of the entire dataframe.
df.describe()
```

```python
#Checking the datatypes of individual columns in the dataframe.
df.info()
```

```python
#Checking if there are any null instances in any columns of the dataframe.
df.isnull().sum()
```

```python
#Now i will check the distribution of the transcations in the dataframe.
df["Class"].value_counts()
#Here 0 represent normal transactions while 1 denotes fraudlent transactions.
#From this output it can be said that the dataset is highly unbalanced.
```

```python
#Seperating the dataset for further analysis
transac_fraud = df[df["Class"] == 1]
transac_legit = df[df["Class"] == 0]
```

```python
print(transac_legit.shape)
print(transac_fraud.shape)
```

```python
transac_legit["Amount"].describe()
```

```python
transac_fraud["Amount"].describe()
```

```python
df.groupby("Class").mean()
```

```python
 #Since the original dataframe is biased, so in order to avoid undersampling, need to select same instances as a random sample from legit transactions.
 unbiased_transac_legit = transac_legit.sample(n = 492)
```

```python
unbiased_transac_legit.head(3)
```

```python
unbiased_transac_legit.shape
```

```python
#Joining the two dataframes to create a new unbiased dataframe free from undersampling
df_new = pd.concat([unbiased_transac_legit, transac_fraud], axis = 0)
```

```python
df_new.shape
```

```python
df_new.groupby("Class").mean()
```

```python
#Removing labels from feature dataframe and adding same to a new label dataframe
feature = df_new.drop("Class", axis = 1)
label = df_new["Class"]
```

```python
print(feature.shape)
print(label.shape)
```

```python
#Splitting the dataframe into child dataframe for traninng and testing.
feature_train, feature_test, label_train, label_test = train_test_split(feature, label, test_size= 0.1, random_state = 1, stratify = label)
```

```python
print(feature_train.shape)
print(feature_test.shape)
print(label_train.shape)
print(label_test.shape)
```

```python
#Selecting model for further classification analysis
classifier = LogisticRegression()
classifier.fit(feature_train, label_train)
```

```python
#Running the model on training data and checking accuracy of the same
predict_train = classifier.predict(feature_train)
train_accuracy = accuracy_score(predict_train, label_train)
print(round(train_accuracy, 2)*100)
```

```python
#Running the model on test data and checking accuracy of the same
predict_test = classifier.predict(feature_test)
test_accuracy = accuracy_score(predict_test, label_test)
print(round(test_accuracy, 2)*100)
```
