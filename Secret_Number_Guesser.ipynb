```python
import random
```

First I will be creating a game where i will be guessing the secret number generated by the computer -

```python
def guess_num(n):
  random_num = random.randint(1,n)
  guess_num = 0
  while guess_num != random_num:
    guess_num = int(input(f"Guess the random number between 1 and {n}"))
    if guess_num > random_num:
      print("Your number is larger than the secret number, guess low!!")
    elif guess_num < random_num:
      print("Your number is smaller than the secret number, guess high!!")
  print(f"You've correctly guessed the secret number: {random_num}")
```

Now, I will create a game where computer will be guessing a secret number that i will think of -

```python
def comp_guess(m):
  low = 1
  high = m
  feed = ""
  while feed != "c":
    guess = random.randint(low, high)
    feed = input(f"If The number {guess} is too high: Type (H), \ if the guess is too low: type (L) \ and if the guess is perfect: type (c)").lower()
    if feed == "h":
      high = guess - 1
    elif feed == "l":
      low = low + 1
  print(f"yoyo! you have guessed the correct number {guess}")
```

```python
comp_guess(10)
```
